### Data Scientist 

### Task and Data 

Using two input CSVs, create a Spark Scala/PySpark/Dask application (you can use any other distributed framework for data processing) which will process input data and create and store a new CSV with insights (what we call gleans).

The app with the code should be pushed to GitHub/GitLab or any other git based platform.

#### Input CSVs structure:

- invoice
 	invoice_id - Invoice UUID
	invoice_date - Issue date
	due_date - Due date
	period_start_date - Start date (if any) of service dates that invoice covers
	period_end_date - End date (if any) of service dates that invoice covers
	total_amount - Billed Amount
	canonical_vendor_id - Vendor UUID

- line_item
 	invoice_id - Invoice UUID
	line_item_id - Raw Line Item UUID
	period_start_date - Start date (if any) of service dates that raw line item covers
	period_end_date - End date (if any) of service dates that raw line item covers
	total_amount - Billed Amount
	canonical_line_item_id - Canonical line item UUID
    
    
#### Output CSV structure:

	glean_id - Glean UUID
	glean_date - Date when glean was triggered
	glean_text - Glean Text
	glean_type - Enum (vendor_not_seen_in_a_while, accrual_alert, large_month_increase_mtd, no_invoice_received)
	glean_location - Enum (invoice or vendor)
	invoice_id - Invoice UUID
	canonical_vendor_id - Vendor UUID    

#### Additional notes:
- 1:M relationship between invoice and line_item
- 1:M relationship between line_item_id and canonical_line_item_id (i.e. different raw line items can be mapped to a single canonical line item)

#### Gleans (insights) to be generated and backfilled in the past:

##### vendor_not_seen_in_a_while

	logic:
	1) don't trigger if invoice received from the vendor for the first time
	2) trigger if invoice received from the vendor and it's > 90 days since last `invoice_date`

	glean_text (text of the glean to be created):
		First new bill in [count_of_months_since_last_invoice] months from vendor [canonical_vendor_id]

	glean_location:
		invoice (this glean should be created on an invoice level)

##### accrual_alert

	logic:
	1) trigger if `period_end_date` for invoice or any line item > 90 days after `invoice_date`
	   If there are multiple end dates, pick the last one.

	glean_text (text of the glean to be created):
		Line items from vendor [canonical_vendor_id] in this invoice cover future periods (through [period_end_date])

	glean_location:
		invoice (this glean should be created on an invoice level)

##### large_month_increase_mtd

	logic:
	1) trigger if monthly spend > $10K and it increased > 50% of average spend over last 12 months. If monthly spend is less than $10K, > 200%. If less than $1K, > 500%. If less than $100, don't trigger the glean. Spend is sum of invoice `total_amount`.


	glean_text (text of the glean to be created):
		Monthly spend with [canonical_vendor_id] is $x (x%) higher than average

	glean_location:
		vendor (this glean should be created on a vendor level)

##### no_invoice_received

	logic:
	1) trigger if vendor sends invoice(s) either on MONTHLY basis (1 or more per month) or QUARTERLY basis (1 per quarter).

		1.1) MONTHLY case: trigger if there were 3 consecutive months with invoices received from vendor but there are no invoices received in current month.
			 Start triggering the glean from the day when vendor usually sends the invoice (you need to count day frequency). Keep triggering the glean till the end of the current month or until the day when new invoice received.
			 If there are multiple days based on frequency count, pick the earliest one.

		1.2) QUARTERLY case: trigger if there were 2 consecutive quarters with invoices received from vendor but there are no invoices received in current quarter.
			 Start triggering the glean from the day when vendor usually sends the invoice (you need to count day frequency). Keep triggering the glean till the end of the current month of the quarter or until the day when new invoice received.
			 If there are multiple days based on frequency count, pick the earliest one.


	glean_text (text of the glean to be created):
		[canonical_vendor_id] generally charges between on [most_frequent_day_number] day of each month invoices are sent. On [date], an invoice from [canonical_vendor_id] has not been received

	glean_location:
		vendor (this glean should be created on a vendor level)

### vendor_not_seen_in_a_while

	logic:
	1) don't trigger if invoice received from the vendor for the first time
	2) trigger if invoice received from the vendor and it's > 90 days since last `invoice_date`

	glean_text (text of the glean to be created):
		First new bill in [count_of_months_since_last_invoice] months from vendor [canonical_vendor_id]

	glean_location:
		invoice (this glean should be created on an invoice level)


```python
vendor_not_seen_in_a_while_gleans = spark.sql("""
select 
    uuid() as glean_id,
    invoice_date as glean_date,
--    lag_invoice_date,
    concat(
        'First new bill in ', 
        CAST(months_between(invoice_date, lag_invoice_date) as int),
        ' months from vendor ', 
        canonical_vendor_id
     ) 
as glean_text,     
    'vendor_not_seen_in_a_while' as glean_type,
    "invoice" as glean_location, 
    invoice_id,
    canonical_vendor_id
from (
    select 
        *,
        LAG(i.invoice_date, 1, '') OVER (PARTITION BY i.canonical_vendor_id ORDER BY i.invoice_date) 
            as lag_invoice_date
    from invoice i
) invoice_with_lag 
where datediff(invoice_date, lag_invoice_date) > 90

""").cache()
vendor_not_seen_in_a_while_gleans.limit(1).toPandas().T
```

### accrual_alert

	logic:
	1) trigger if `period_end_date` for invoice or any line item > 90 days after `invoice_date`
	   If there are multiple end dates, pick the last one.

	glean_text (text of the glean to be created):
		Line items from vendor [canonical_vendor_id] in this invoice cover future periods (through [period_end_date])

	glean_location:
		invoice (this glean should be created on an invoice level)

```python
accural_alert_gleans = spark.sql("""
    select
        uuid() as glean_id,
        invoice_date as glean_date,
        concat(
            'Line items from vendor ', 
            canonical_vendor_id,
            'in this invoice cover future periods (through ', 
            max_period_end_date, ')'
        ) 
        as glean_text,     
        'accrual_alert' as glean_type,
        "invoice" as glean_location, 
        invoice_id,
        canonical_vendor_id
    from (
        select 
            i.invoice_id,
            i.invoice_date, 
            i.canonical_vendor_id, 
            max(l.period_end_date) as max_period_end_date
        from invoice i 
        left join line_item l 
        on i.invoice_id = l.invoice_id   
        group by 
            i.invoice_id,
            i.invoice_date, 
            i.canonical_vendor_id
    ) vendor_and_line    
    where datediff(max_period_end_date, invoice_date) > 90
""").cache()

accural_alert_gleans.limit(1).toPandas().T
```

### large_month_increase_mtd

	logic:
	1) trigger if monthly spend > $10K and it increased > 50% of average spend over last 12 months. If monthly spend is less than $10K, > 200%. If less than $1K, > 500%. If less than $100, don't trigger the glean. Spend is sum of invoice `total_amount`.


	glean_text (text of the glean to be created):
		Monthly spend with [canonical_vendor_id] is $x (x%) higher than average

	glean_location:
		vendor (this glean should be created on a vendor level)


- glean_id - Glean UUID
- glean_date - Date when glean was triggered
- glean_text - Glean Text
- glean_type - Enum (vendor_not_seen_in_a_while, accrual_alert, large_month_increase_mtd, no_invoice_received)
- glean_location - Enum (invoice or vendor)
- invoice_id - Invoice UUID
- canonical_vendor_id - Vendor UUID  

```python
## large_month_increase_mtd (Create monthly_spend vs prior avg)

monthly_spend_vs_prior_avg = spark.sql("""
WITH monthly_total as (
        SELECT 
         i.canonical_vendor_id, 
         sum(i.total_amount) as monthly_spend,
         max(i.invoice_date) as max_invoice_date,
         cast(month(i.invoice_date) as int) as month, 
         cast(year(i.invoice_date) as int) as year,         
         cast(month(i.invoice_date) as int) +
         cast(year(i.invoice_date) as int) * 12 as monthnum
FROM invoice i
GROUP BY i.canonical_vendor_id,
         month(i.invoice_date), 
         year(i.invoice_date)
ORDER BY year(i.invoice_date),
         cast(month(i.invoice_date) as int)
         
)
SELECT 
    monthly_total.monthnum,
    monthly_total.canonical_vendor_id,
    monthly_total.monthly_spend,
    monthly_total.max_invoice_date,
    AVG(preceding_months.monthly_spend) as TTM_avg,
    COUNT(preceding_months.monthly_spend) as average_count
FROM
    monthly_total
LEFT JOIN
    monthly_total preceding_months ON (
        preceding_months.monthnum BETWEEN monthly_total.monthnum - 12 AND monthly_total.monthnum - 1
        AND preceding_months.canonical_vendor_id = monthly_total.canonical_vendor_id
    ) 

GROUP BY
    monthly_total.monthly_spend,
    monthly_total.monthnum, 
    monthly_total.canonical_vendor_id,
    monthly_total.max_invoice_date
ORDER BY
    monthly_total.canonical_vendor_id,
    monthly_total.monthnum
""")
monthly_spend_vs_prior_avg.registerTempTable("monthly_spend_vs_prior_avg")

monthly_spend_vs_prior_avg.limit(10).toPandas()
```

```python
# large_month_increase_mtd
mtd_increase_glean = spark.sql("""
select 
        uuid() as glean_id,
        max_invoice_date as glean_date,
        concat(
            'Monthly spend with ',
            canonical_vendor_id,
            'is $x (x%) higher than average'
        ) 
        as glean_text,     
        'large_month_increase_mtd' as glean_type,
        'vendor' as glean_location, 
        null as invoice_id,
        canonical_vendor_id
from monthly_spend_vs_prior_avg 
WHERE 
    (monthly_spend between 100 and 1000 and monthly_spend /TTM_avg > 6)
    or (monthly_spend between 1001 and 10000 and monthly_spend /TTM_avg > 3)
    or (monthly_spend > 10001 and monthly_spend /TTM_avg > 1.5)
""").cache()
mtd_increase_glean.limit(5).toPandas()
```

### no_invoice_received

	logic:
	1) trigger if vendor sends invoice(s) either on MONTHLY basis (1 or more per month) or QUARTERLY basis (1 per quarter).

		1.1) MONTHLY case: trigger if there were 3 consecutive months with invoices received from vendor but there are no invoices received in current month.
			 Start triggering the glean from the day when vendor usually sends the invoice (you need to count day frequency). Keep triggering the glean till the end of the current month or until the day when new invoice received.
			 If there are multiple days based on frequency count, pick the earliest one.

		1.2) QUARTERLY case: trigger if there were 2 consecutive quarters with invoices received from vendor but there are no invoices received in current quarter.
			 Start triggering the glean from the day when vendor usually sends the invoice (you need to count day frequency). Keep triggering the glean till the end of the current month of the quarter or until the day when new invoice received.
			 If there are multiple days based on frequency count, pick the earliest one.


	glean_text (text of the glean to be created):
		[canonical_vendor_id] generally charges between on [most_frequent_day_number] day of each month invoices are sent. On [date], an invoice from [canonical_vendor_id] has not been received

	glean_location:
		vendor (this glean should be created on a vendor level)

```python
# Relevant_date_range for create date dimension table

date_range_results = spark.sql("""
SELECT 
    min(invoice_date) as min_date, 
    max(invoice_date) as max_date
FROM invoice
""")
min_date, max_date = date_range_results.take(1)[0]

end_of_next_quarter = (
    pd.to_datetime(max_date) 
    + pd.tseries.offsets.DateOffset(months=3) 
    + pd.tseries.offsets.QuarterEnd()
)
relevant_date_range = pd.date_range(min_date, end_of_next_quarter)
```

```python
relevant_date_range
```

```python
from datetime import timedelta
day_of_quarter = (relevant_date_range - (relevant_date_range - pd.tseries.offsets.QuarterEnd())).days
```

```python
# Create date dimension table
pd_all_days = pd.DataFrame(
    {
        "date":relevant_date_range.date, 
        'day_of_quarter':day_of_quarter,
        'month_number':relevant_date_range.month + relevant_date_range.year * 12,
        'day_of_month': relevant_date_range.day,
        'quarter_number': (relevant_date_range.month - 1) // 3 + relevant_date_range.year * 4
    }
)

date_dimension = spark.createDataFrame(pd_all_days)
date_dimension.registerTempTable("date_dimension")
```

```python
# day_of_month_frequencies
day_of_month_frequencies = spark.sql("""
SELECT 
    canonical_vendor_id, 
    DAY(invoice_date) as day_of_month, 
    count(*) as frequency
FROM invoice
GROUP by canonical_vendor_id, DAY(invoice_date)
ORDER by canonical_vendor_id, day_of_month
""").cache()
day_of_month_frequencies.registerTempTable("dom_frequencies")
day_of_month_frequencies.limit(30).toPandas()
```

```python
spark.sql("select * from date_dimension").printSchema()
spark.sql("select * from invoice").printSchema()
```

```python
# day_of_quarter_frequencies
day_of_quarter_frequencies = spark.sql("""
SELECT 
    canonical_vendor_id, 
    day_of_quarter, 
    count(*) as frequency
FROM invoice i join date_dimension d on i.invoice_date = d.date

GROUP by canonical_vendor_id, day_of_quarter
ORDER by canonical_vendor_id, day_of_quarter
""").cache()
day_of_quarter_frequencies.registerTempTable("doq_frequencies")
day_of_quarter_frequencies.limit(30).toPandas()
```

```python
# most_frequent_day
most_frequent_day = spark.sql("""
SELECT DISTINCT
    canonical_vendor_id,
    FIRST_VALUE(day_of_month) 
        OVER (
            PARTITION BY canonical_vendor_id 
            ORDER BY frequency DESC, day_of_month ASC
        ) as most_frequent_day
FROM
    dom_frequencies
ORDER BY 
    canonical_vendor_id
""").cache()
most_frequent_day.registerTempTable("most_frequent_day")
most_frequent_day.limit(10).toPandas()
```

```python
# most_frequent_day_quarter
most_frequent_day_quarter = spark.sql("""
SELECT DISTINCT
    canonical_vendor_id,
    FIRST_VALUE(day_of_quarter) 
        OVER (
            PARTITION BY canonical_vendor_id 
            ORDER BY frequency DESC, day_of_quarter ASC
        ) as most_frequent_day
FROM
    doq_frequencies
ORDER BY 
    canonical_vendor_id
""").cache()
most_frequent_day_quarter.registerTempTable("most_frequent_day_quarter")
most_frequent_day_quarter.limit(10).toPandas()
```

```python
# Create invoice_month_count
invoice_month_counts = spark.sql("""
WITH invoice_with_month_number AS (
    SELECT 
        *, 
        YEAR(invoice_date) * 12 + MONTH(invoice_date) as month_number
    FROM 
        invoice
)
SELECT 
    date_dimension.date,
    date_dimension.day_of_month,
    vendors.canonical_vendor_id,
    COUNT(DISTINCT 
        CASE WHEN 
            invoice_with_month_number.month_number = date_dimension.month_number 
        THEN 
            invoice_with_month_number.month_number
        END
    ) as invoice_count_for_current_month,
    COUNT(DISTINCT 
        CASE WHEN 
            invoice_with_month_number.month_number 
                BETWEEN date_dimension.month_number - 3 and date_dimension.month_number - 1
        THEN 
            invoice_with_month_number.month_number
        END
    ) as invoice_count_for_preceding_months
FROM
    date_dimension
JOIN
    (SELECT DISTINCT canonical_vendor_id FROM invoice) vendors
JOIN
    invoice_with_month_number 
        ON (
            invoice_with_month_number.canonical_vendor_id = vendors.canonical_vendor_id
            AND invoice_with_month_number.invoice_date <= date_dimension.date -- only select older invoices
            -- Below is stictly for performance
            AND invoice_with_month_number.month_number 
                BETWEEN date_dimension.month_number - 3 and date_dimension.month_number                 
        )
GROUP BY
    date_dimension.date, date_dimension.day_of_month, vendors.canonical_vendor_id
""")
invoice_month_counts.createOrReplaceTempView("invoice_month_counts")
```

```python
# Create date dimension table
pd_all_days = pd.DataFrame(
    {
        "date":relevant_date_range, 
        'day_of_quarter':day_of_quarter,
        'month_number':relevant_date_range.month + relevant_date_range.year * 12,
        'day_of_month': relevant_date_range.day,
        'quarter_number': (relevant_date_range.month - 1) // 3 + relevant_date_range.year * 4
    }
)

date_dimension = spark.createDataFrame(pd_all_days)
date_dimension.registerTempTable("date_dimension")
```

```python
# Create invoice_quarter_count
invoice_quarter_counts = spark.sql("""
WITH invoice_with_quarter_number AS (
    SELECT 
        *,              
        YEAR(invoice_date) * 4 + FLOOR((MONTH(invoice_date) - 1) / 3) as quarter_number              
    FROM 
        invoice
)
SELECT 
    date_dimension.date,
    date_dimension.day_of_quarter,
    vendors.canonical_vendor_id,
    COUNT(
        CASE WHEN 
            invoice_with_quarter_number.quarter_number = date_dimension.quarter_number 
        THEN 
            invoice_with_quarter_number.quarter_number
        END
    ) as invoice_count_for_current_quarter,
    COUNT(
        CASE WHEN 
            invoice_with_quarter_number.quarter_number = date_dimension.quarter_number - 1
        THEN 
            invoice_with_quarter_number.quarter_number 
        END
    ) as invoice_count_for_preceding_quarter_1,
    COUNT(
        CASE WHEN 
            invoice_with_quarter_number.quarter_number = date_dimension.quarter_number - 2
        THEN 
            invoice_with_quarter_number.quarter_number 
        END
    ) as invoice_count_for_preceding_quarter_2
FROM
    date_dimension
JOIN
    (SELECT DISTINCT canonical_vendor_id FROM invoice) vendors
JOIN
    invoice_with_quarter_number 
        ON (
            invoice_with_quarter_number.canonical_vendor_id = vendors.canonical_vendor_id
            AND invoice_with_quarter_number.invoice_date <= date_dimension.date -- only select older invoices
            -- Below is stictly for performance
            AND invoice_with_quarter_number.quarter_number 
                BETWEEN date_dimension.quarter_number - 2 and date_dimension.quarter_number                 
        )
GROUP BY
    date_dimension.date, date_dimension.day_of_quarter, vendors.canonical_vendor_id
""")
invoice_quarter_counts.createOrReplaceTempView("invoice_quarter_counts")
```

```python
no_invoice_received_month_glean = spark.sql("""
SELECT 
        uuid() as glean_id,
        invoice_month_counts.date as glean_date,
        concat(
            invoice_month_counts.canonical_vendor_id , 
            ' generally charges between on ', 
            most_frequent_day.most_frequent_day,
            ' day of each month invoices are sent. On ',
            date,
            ', an invoice from ',
            invoice_month_counts.canonical_vendor_id,
            'has not been received'
        ) 
        as glean_text,     
        'no_invoice_received' as glean_type,
        'vendor' as glean_location, 
        null as invoice_id,
        invoice_month_counts.canonical_vendor_id  
FROM 
    invoice_month_counts 
JOIN
    most_frequent_day 
    ON (invoice_month_counts.canonical_vendor_id = most_frequent_day.canonical_vendor_id)
WHERE
    invoice_count_for_current_month = 0
    AND invoice_count_for_preceding_months = 3
    AND invoice_month_counts.day_of_month > most_frequent_day.most_frequent_day
""").cache()
no_invoice_received_month_glean.limit(5).toPandas()
```

```python
no_invoice_received_quarter_glean = spark.sql("""
SELECT 
        uuid() as glean_id,
        invoice_quarter_counts.date as glean_date,
        concat(
            invoice_quarter_counts.canonical_vendor_id , 
            ' generally charges between on ', 
            most_frequent_day_quarter.most_frequent_day,
            ' day of each quarter invoices are sent. On ',
            date,
            ', an invoice from ',
            invoice_quarter_counts.canonical_vendor_id,
            'has not been received'
        ) 
        as glean_text,     
        'no_invoice_received' as glean_type,
        'vendor' as glean_location, 
        null as invoice_id,
        invoice_quarter_counts.canonical_vendor_id 
FROM 
    invoice_quarter_counts 
JOIN
    most_frequent_day_quarter 
    ON (invoice_quarter_counts.canonical_vendor_id = most_frequent_day_quarter.canonical_vendor_id)
WHERE
    invoice_count_for_current_quarter = 0
    AND invoice_count_for_preceding_quarter_1 = 1
    AND invoice_count_for_preceding_quarter_2 = 1
    AND invoice_quarter_counts.day_of_quarter > most_frequent_day_quarter.most_frequent_day
""").cache()
no_invoice_received_quarter_glean.limit(5).toPandas()
```
